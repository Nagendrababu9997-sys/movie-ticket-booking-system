```python
class TicketBooking:

    def __init__(self):
        self.tickets = {"movie1": 40, "movie2": 10, "movie3": 0}
        self.combos = {"combo1": 100, "combo2": 150}

    def book_ticket(self):
        email = input("Enter Email: ")
        movie = input("Enter Movie Name: ")
        count = int(input("No of Tickets: "))

        if movie in self.tickets and self.tickets[movie] >= count:
            self.tickets[movie] -= count
            total = count * 200

            food = input("Food (yes/no): ")

            if food == "yes":
                print(self.combos)
                combo = input("Select Combo: ")
                items = int(input("No of Items: "))
                total += self.combos[combo] * items

            print("Booking Successful")
            print("Total Amount =", total)

        else:
            print("Tickets Not Available")

obj = TicketBooking()
obj.book_ticket()
```
